```python
final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
ratings = final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0 : 'Rating Count'})
```

```python
ratings
```

#### Observation:
 1.  When rating between 4.5 - 4.9 ----> Excellent
 2.  When rating between 4.0 - 4.4 ----> Very Good
 3.  When rating between 3.5 - 3.9 ----> Good
 4.  When rating between 2.5 - 3.4 ----> Average
 5.  When rating between 1.8 - 2.4 ----> Poor
 6.  But most of them are rating 0 , ie 2148 are not rated. 

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = [12,6]
sns.barplot(x = 'Aggregate rating', y = 'Rating Count',data = ratings)
```

```python
sns.barplot(x = 'Aggregate rating', y = 'Rating Count',hue = 'Rating color',palette = ['blue','red','orange','yellow','green','green'],data = ratings)
```

#### Observation

1. Not Rated count is very high
2. Max no of rating between 2.5 to 3.4

```python
sns.countplot(x = 'Rating color',data = ratings, palette = ['blue','red','orange','yellow','green','green'])
```
